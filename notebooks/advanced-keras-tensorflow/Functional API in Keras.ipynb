#### Adımlar: 
**1. Adım: Gerekli Kütüphaneleri İçe Aktarın**

Başlamadan önce, gerekli kütüphaneleri içe aktardığınızdan emin olun: TensorFlow ve Keras. Keras, TensorFlow içinde `tensorflow.keras` olarak bulunur. 


````python
# Install TensorFlow
!pip install tensorflow==2.16.2

import tensorflow as tf 
from tensorflow.keras.models import Model 
from tensorflow.keras.layers import Input, Dense 
import warnings
warnings.filterwarnings('ignore', category=UserWarning, module='tensorflow')
````

**Açıklama:**

`!pip install tensorflow==2.16.2` komutu, belirtilen TensorFlow sürümünü yükler.

 `tensorflow`, Python'da makine öğrenimi için kullanılan ana kütüphanedir. 

 `Model`, Functional API ile bir model oluşturmak için kullanılır. 

 `Input` ve `Dense`, modelinizde kullanacağınız katman türleridir. 

 
**Adım 2: Giriş Katmanını Tanımlayın**

Modeliniz için giriş şeklini tanımlayacaksınız. Basitlik açısından, her girişin uzunluğu 20 olan bir vektör olduğu bir veri kümesiyle çalıştığınızı varsayalım. 

 


Translated with DeepL.com (free version)

````python
input_layer = Input(shape=(20,))
print(input_layer)
````

**Açıklama:**

`Input(shape=(20,))`, uzunluğu 20 olan girdi vektörlerini bekleyen bir girdi katmanı oluşturur.

`print(input_layer)`, katman bilgilerini gösterir ve katmanlar hakkında hangi tür bilgileri edinebileceğinizi anlamanıza yardımcı olur.


**3. Adım: Gizli Katmanlar Ekleme**

Ardından, modelinize birkaç gizli katman ekleyeceksiniz. Gizli katmanlar, modelin verilerdeki karmaşık kalıpları öğrenmesine yardımcı olur.


````python
hidden_layer1 = Dense(64, activation='relu')(input_layer) 
hidden_layer2 = Dense(64, activation='relu')(hidden_layer1) 
````

**Açıklama:**

`Dense(64, activation=‘relu’)`, 64 birimden oluşan ve ReLU aktivasyon fonksiyonuna sahip yoğun (tam bağlantılı) bir katman oluşturur. 

Her gizli katman, bir önceki katmanın çıktısını girdi olarak alır. 

**Adım 4: Çıkış Katmanını Tanımlayın**

Son olarak, çıkış katmanını tanımlayacaksınız. İkili sınıflandırma problemi üzerinde çalıştığınızı varsayalım; bu durumda çıkış katmanı, sigmoid aktivasyon fonksiyonuna sahip tek bir birime sahip olacaktır. 


````python
output_layer = Dense(1, activation='sigmoid')(hidden_layer2) 
````

**Açıklama:**

`Dense(1, activation=‘sigmoid’)`, ikili sınıflandırma için uygun, 1 birimlik ve sigmoid aktivasyon fonksiyonuna sahip bir yoğun katman oluşturur. 

**5. Adım: Modeli Oluşturma**

Şimdi, giriş ve çıkış katmanlarını belirleyerek modeli oluşturacaksınız. 


````python
model = Model(inputs=input_layer, outputs=output_layer)
model.summary()
````

**Açıklama:**

`Model(inputs=input_layer, outputs=output_layer)`, gizli katmanlar aracılığıyla giriş katmanını çıkış katmanına bağlayan bir Keras modeli oluşturur. 

`model.summary()`, katmanları, şekillerini ve parametre sayısını gösteren bir model özeti sunar. Bu, model mimarisini anlamanıza yardımcı olur.


**6. Adım: Modeli Derleyin**

Modeli eğitmeden önce derlemeniz gerekir. Kayıp fonksiyonunu, optimizeri ve değerlendirme ölçütlerini belirleyeceksiniz. 


````python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
````

**Açıklama:**

`optimizer=‘adam’`, sinir ağlarının eğitilmesinde yaygın olarak kullanılan Adam optimizörünü belirtir. 

`loss=‘binary_crossentropy’`, ikili sınıflandırma problemleri için kayıp fonksiyonunu belirtir. 

`metrics=[‘accuracy’]`, Keras'a eğitim sırasında modeli doğruluk oranını kullanarak değerlendirmesini söyler. 

**7. Adım: Modeli Eğitin**

Artık modeli eğitim verilerini kullanarak eğitebilirsiniz. Bu örnek için, `X_train`'in eğitim girdi verileriniz ve `y_train`'in karşılık gelen etiket olduğunu varsayalım. 


````python
# Example data (in practice, use real dataset) 

import numpy as np 
X_train = np.random.rand(1000, 20) 
y_train = np.random.randint(2, size=(1000, 1)) 
model.fit(X_train, y_train, epochs=10, batch_size=32) 
````

**Açıklama:**

`X_train` ve `y_train`, gerçek eğitim verilerinizin yerine geçen yer tutuculardır. 

`model.fit`, modeli belirtilen epoch sayısı ve parti boyutu ile eğitir. 

**8. Adım: Modeli Değerlendirme**

Eğitimden sonra, modelin performansını görmek için modeli test verileri üzerinde değerlendirebilirsiniz. 


````python
# Example test data (in practice, use real dataset) 

X_test = np.random.rand(200, 20) 
y_test = np.random.randint(2, size=(200, 1)) 
loss, accuracy = model.evaluate(X_test, y_test) 
print(f'Test loss: {loss}') 
print(f'Test accuracy: {accuracy}') 
````

**Açıklama:**

`model.evaluate`, test verileri üzerinde modelin kayıp değerini ve doğruluğunu hesaplar. 

`X_test` ve `y_test`, gerçek test verilerinizin yerine geçen yer tutuculardır. 


### Dropout ve Toplu Normalleştirme

Uygulama alıştırmasına geçmeden önce, sinir ağlarının performansını artırmak için sıklıkla kullanılan iki önemli tekniği kısaca ele alalım: **Dropout Katmanları** ve **Toplu Normalleştirme**.

#### Dropout Katmanları

Dropout, sinir ağlarında aşırı uyumun önlenmesine yardımcı olan bir düzenleme tekniğidir. Eğitim sırasında Dropout, her güncelleme döngüsünde giriş birimlerinin bir kısmını rastgele sıfıra ayarlar. Bu, modelin belirli nöronlara aşırı bağımlı hale gelmesini önler ve ağın, görülmemiş verilere daha iyi genelleştirilebilen daha sağlam özellikleri öğrenmesini teşvik eder.

**Önemli noktalar:**
- Dropout yalnızca eğitim sırasında uygulanır, çıkarım sırasında uygulanmaz.
- Dropout oranı, atılacak nöronların oranını belirleyen bir hiperparametredir.


#### Toplu Normalleştirme

Toplu Normalleştirme, sinir ağlarının eğitim kararlılığını ve hızını artırmak için kullanılan bir tekniktir. Verileri yeniden merkezleyip yeniden ölçeklendirerek önceki katmanın çıktısını normalleştirir; bu da öğrenme sürecinin stabilize edilmesine yardımcı olur. İç kovaryans kaymasını (katman girdilerinin dağılımındaki değişiklikler) azaltarak, toplu normalleştirme modelin daha yüksek öğrenme hızları kullanmasına olanak tanır; bu da genellikle yakınsamayı hızlandırır.

**Önemli Noktalar:**

- Toplu normalizasyon, her katmana gelen girdileri ortalama sıfır ve varyans bir olacak şekilde normalize ederek çalışır.
- Hem eğitim hem de çıkarım sırasında uygulanır, ancak bu iki aşamada davranışı biraz farklılık gösterir.
- Toplu

**Example of adding a Dropout layer in Keras:**


````python
from tensorflow.keras.layers import Dropout, Dense, Input
from tensorflow.keras.models import Model

# Define the input layer
input_layer = Input(shape=(20,))

# Add a hidden layer
hidden_layer = Dense(64, activation='relu')(input_layer)

# Add a Dropout layer
dropout_layer = Dropout(rate=0.5)(hidden_layer)

# Add another hidden layer after Dropout
hidden_layer2 = Dense(64, activation='relu')(dropout_layer)

# Define the output layer
output_layer = Dense(1, activation='sigmoid')(hidden_layer2)

# Create the model
model = Model(inputs=input_layer, outputs=output_layer)

# Summary of the model
model.summary()
````

**Example of adding Batch Normalization in Keras:**


````python
from tensorflow.keras.layers import BatchNormalization, Dense, Input
from tensorflow.keras.models import Model

# Define the input layer
input_layer = Input(shape=(20,))

# Add a hidden layer
hidden_layer = Dense(64, activation='relu')(input_layer)

# Add a BatchNormalization layer
batch_norm_layer = BatchNormalization()(hidden_layer)

# Add another hidden layer after BatchNormalization
hidden_layer2 = Dense(64, activation='relu')(batch_norm_layer)

# Define the output layer
output_layer = Dense(1, activation='sigmoid')(hidden_layer2)

# Create the model
model = Model(inputs=input_layer, outputs=output_layer)

# Summary of the model
model.summary()
````
