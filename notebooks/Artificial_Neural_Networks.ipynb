## Bir Sinir Ağı Oluşturun

Öncelikle ağın yapısını resmi olarak tanımlayarak başlayalım.

```python
n = 2 # giriş sayısı
num_hidden_layers = 2 # gizli katman sayısı
m = [2, 2] # her gizli katmandaki düğüm sayısı
num_nodes_output = 1 # çıkış katmanındaki düğüm sayısı
```

Ağ yapısını tanımladığımıza göre, şimdi ağdaki ağırlıkları ve sapmaları rastgele sayılarla başlatalım. Ağırlıkları ve sapmaları rastgele sayılarla başlatabilmek için **Numpy** kütüphanesini içe aktarmamız gerekecek.

```python
import numpy as np 

num_nodes_previous = n  # Giriş katmanındaki (veya bir önceki katmandaki) nöron sayısı

network = {}  # Sinir ağını saklayacağımız boş sözlük (dictionary)

# Gizli katmanlar + çıktı katmanı için döngü başlatıyoruz
# Çıktı katmanını dahil etmek için gizli katman sayısına 1 ekliyoruz
for layer in range(num_hidden_layers + 1): 
    
    # Eğer mevcut katman indeksi gizli katman sayısına eşitse
    # Bu, artık çıktı katmanında olduğumuz anlamına gelir
    if layer == num_hidden_layers:
        layer_name = 'output'              # Katman adını "output" olarak belirle
        num_nodes = num_nodes_output       # Çıktı katmanındaki nöron sayısını ata
    else:
        # Aksi durumda bu bir gizli katmandır
        layer_name = 'layer_{}'.format(layer + 1)  # Örn: layer_1, layer_2 ...
        num_nodes = m[layer]                      # Bu katmandaki nöron sayısını listeden al
    
    # Mevcut katman için sözlük oluştur
    network[layer_name] = {}
    
    # Katmandaki her nöron için ağırlık ve bias değerlerini rastgele başlat
    for node in range(num_nodes):
        
        node_name = 'node_{}'.format(node + 1)  # Örn: node_1, node_2 ...
        
        network[layer_name][node_name] = {
            # Ağırlıklar:
            # Önceki katmandaki her nöronla bağlantı olduğu için
            # ağırlık sayısı = önceki katmandaki nöron sayısı
            'weights': np.around(
                np.random.uniform(size=num_nodes_previous), 
                decimals=2
            ),
            
            # Bias:
            # Her nöron için tek bir bias değeri
            'bias': np.around(
                np.random.uniform(size=1), 
                decimals=2
            ),
        }
    
    # Bir sonraki katman için
    # "önceki katmandaki nöron sayısını" güncelle
    num_nodes_previous = num_nodes

# Oluşturulan sinir ağını ekrana yazdır
print(network)
```

Harika! Yukarıdaki kod sayesinde, herhangi bir sayıda gizli katmana ve her katmanda herhangi bir sayıda düğüme sahip herhangi bir ağın ağırlıklarını ve sapmalarını başlatabiliyoruz. Ancak bu kodu bir fonksiyona koyalım ki, bir sinir ağı oluşturmak istediğimiz her zaman bu kodu tekrarlı olarak çalıştırabilelim.

```python
def initialize_network(num_inputs, num_hidden_layers, num_nodes_hidden, num_nodes_output):
    
    num_nodes_previous = num_inputs # number of nodes in the previous layer

    network = {}
    
    # loop through each layer and randomly initialize the weights and biases associated with each layer
    for layer in range(num_hidden_layers + 1):
        
        if layer == num_hidden_layers:
            layer_name = 'output' # name last layer in the network output
            num_nodes = num_nodes_output
        else:
            layer_name = 'layer_{}'.format(layer + 1) # otherwise give the layer a number
            num_nodes = num_nodes_hidden[layer] 
        
        # initialize weights and bias for each node
        network[layer_name] = {}
        for node in range(num_nodes):
            node_name = 'node_{}'.format(node+1)
            network[layer_name][node_name] = {
                'weights': np.around(np.random.uniform(size=num_nodes_previous), decimals=2),
                'bias': np.around(np.random.uniform(size=1), decimals=2),
            }
    
        num_nodes_previous = num_nodes

    return network # return the network
```

#### Aşağıdaki özelliklere sahip bir ağ oluşturmak için *initialize_network* fonksiyonunu kullanın:

1. 5 giriş alır
2. Üç gizli katmana sahiptir
3. Birinci katmanda 3 düğüm, ikinci katmanda 2 düğüm ve üçüncü katmanda 3 düğüm bulunur
4. Çıkış katmanında 1 düğüm bulunur

Ağı **small_network** olarak adlandırın.

```python
small_network = initialize_network(5, 3, [3, 2, 3], 1)
```

## Her Düğümde Ağırlıklı Toplamı Hesapla

Her düğümdeki ağırlıklı toplam, girdilerin ve ağırlıkların nokta çarpımı artı sapma olarak hesaplanır. Öyleyse, tam olarak bunu yapan *compute_weighted_sum* adlı bir fonksiyon oluşturalım.

```python
def compute_weighted_sum(inputs, weights, bias):
    return np.sum(inputs * weights) + bias
```

Şimdi **small_network**'e besleyebileceğimiz 5 girdi oluşturalım.

```python
from random import seed
import numpy as np

np.random.seed(12)
inputs = np.around(np.random.uniform(size=5), decimals=2)

print('Ağa giren veriler şunlardır:{}'.format(inputs))
```

```python
node_weights = small_network['layer_1']['node_1']['weights']
node_bias = small_network['layer_1']['node_1']['bias']

weighted_sum = compute_weighted_sum(inputs, node_weights, node_bias)
print('Gizli katmandaki ilk düğümdeki ağırlıklı toplam şöyledir: {}'.format(np.around(weighted_sum[0], decimals=4)))
```

## Hesaplama Düğümü Etkinleştirme

Her düğümün çıktısının, ağırlıklı toplamın doğrusal olmayan bir dönüşümü olduğunu hatırlayalım. Bu eşleme için aktivasyon fonksiyonları kullanıyoruz. Burada aktivasyon fonksiyonu olarak sigmoid fonksiyonunu kullanalım. Öyleyse, girdi olarak ağırlıklı bir toplam alan ve sigmoid fonksiyonunu kullanarak girdinin doğrusal olmayan dönüşümünü döndüren bir fonksiyon tanımlayalım.

```python
def node_activation(weighted_sum):
    return 1.0 / (1.0 + np.exp(-1 * weighted_sum))
```

#### İlk gizli katmandaki ilk düğümün çıktısını hesaplamak için *node_activation* fonksiyonunu kullanın.

```python
node_output  = node_activation(compute_weighted_sum(inputs, node_weights, node_bias))
print('The output of the first node in the hidden layer is {}'.format(np.around(node_output[0], decimals=4)))
```

## İleri Yayılım

Tahmin yapabilen bir sinir ağı oluşturmanın son aşaması, her şeyi bir araya getirmektir. Bu nedenle, ağdaki her düğüme *compute_weighted_sum* ve *node_activation* fonksiyonlarını uygulayan, verileri çıktı katmanına kadar ileten ve çıktı katmanındaki her düğüm için bir tahmin üreten bir fonksiyon oluşturalım.

Bunu aşağıdaki prosedürle gerçekleştireceğiz:

1. Giriş katmanını ilk gizli katmanın girişi olarak kullanarak başlayın.
2. Mevcut katmanın düğümlerindeki ağırlıklı toplamı hesaplayın.
3. Mevcut katmanın düğümlerinin çıktısını hesaplayın.
4. Mevcut katmanın çıktısını bir sonraki katmanın girişi olarak ayarlayın.
5. Ağdaki bir sonraki katmana geçin.
6. Çıkış katmanının çıktısını hesaplayana kadar 2-5 arasındaki adımları tekrarlayın.

```python
def forward_propagate(network, inputs):
    
    layer_inputs = list(inputs) # start with the input layer as the input to the first hidden layer
    
    for layer in network:
        
        layer_data = network[layer]
        
        layer_outputs = [] 
        for layer_node in layer_data:
        
            node_data = layer_data[layer_node]
        
            # compute the weighted sum and the output of each node at the same time 
            node_output = node_activation(compute_weighted_sum(layer_inputs, node_data['weights'], node_data['bias']))
            layer_outputs.append(np.around(node_output[0], decimals=4))
            
        if layer != 'output':
            print('The outputs of the nodes in hidden layer number {} is {}'.format(layer.split('_')[1], layer_outputs))
    
        layer_inputs = layer_outputs # set the output of this layer to be the input to next layer

    network_predictions = layer_outputs
    return network_predictions
```

#### Küçük ağımızın tahminini hesaplamak için *forward_propagate* fonksiyonunu kullanın

```python
predictions = forward_propagate(small_network, inputs)
print('The predicted value by the network for the given input is {}'.format(np.around(predictions[0], decimals=4)))
```
