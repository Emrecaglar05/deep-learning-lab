__Sigmoid Aktivasyon Fonksiyonu__
* Sigmoid fonksiyonu, giriş değerlerini 0 ile 1 arasında bir aralığa dönüştürür.
* "S şeklinde" bir eğri üretir, bu da onu olasılıksal çıktılar veya ikili sınıflandırma görevleri için uygun hale getirir.

* Alanı boyunca (-∞, +∞) sürekli, türevlenebilir ve monoton olarak artar.

* Türevinin hesaplanması basittir: σ′(z)=σ(z)⋅(1−σ(z)). Bu, geri yayılım için önemlidir.

* Olasılıkların gerekli olduğu çıktı katmanları için kullanışlıdır (örneğin, lojistik regresyon).

* Doğrusal olmayan karar sınırlarını etkinleştirerek sinir ağlarının karmaşık ilişkileri öğrenmesini sağlar.

__Sınırlamalar__

* Kaybolan gradyan probleminden muzdariptir: Büyük pozitif veya negatif girdiler için gradyan sıfıra yaklaşır ve derin ağlarda öğrenmeyi yavaşlatır.

* Üstel işlem içerdiği için hesaplama açısından pahalıdır.

__ReLU (Doğrulanmış Doğrusal Birim) Aktivasyon Fonksiyonu__

* ReLU, f(x)=max⁡(0,x)f(x)=max(0,x) kullanarak sıfır ve giriş değerinin maksimumunu verir.

* Doğrusal değildir ancak sigmoidden daha basittir; negatif girdiler için sıfır, pozitif girdiler için giriş değerini verir.

* Türev basittir: x>0 için f′(x)=1 ve x≤0 için f′(x)=0.

* Pozitif girdiler için sabit bir gradyan (1) sağlayarak kaybolan gradyan sorununu hafifletir ve derin ağlarda daha hızlı yakınsama sağlar.

* Basitliği nedeniyle hesaplama açısından verimlidir.

__Sınırlamalar__

* Negatif girdiler nedeniyle nöronların sürekli olarak sıfır çıktısı verdiği "ölü nöronlara" yol açabilir. Bu, Leaky ReLU gibi varyantlar kullanılarak hafifletilebilir.

## Import required libraries
Let's start with importing the required libraries: Numpy , for matrix calculations
and matplotlib for visualization


```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Sigmoid aktivasyon fonksiyonunu tanımlayın

```python
# Sigmoid function and its derivatives
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

#### Sigmoid türevi, yani "sigmoid gradyanı"nı tanımlayın

```python
def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

#### ReLU fonksiyonunu tanımlayın

```python
def relu(z):
    return np.maximum(0, z)
```

#### Define the ReLU derivative (ReLU gradient)

#### Define the number of input values
Generate 400 synthetic values between -10 and 10. The gradient would be calculated for each of these values.


```python
# Generate a range of input values
z = np.linspace(-10, 10, 400)
```

Next, let's compute the derivatives for Sigmoid and Relu activations and assign the values to 
**sigmoid_grad** and **relu_grad** respectively 


```python
### type your answer here
```

Double-click __here__ for the solution.
<!-- The correct answer is:

relu_grad = relu_derivative(z)
-->


Awesome! So now, let's visualize the sigmoid and relu activations and their gradients by plotting their output values


```python
# Plot the activation functions
plt.figure(figsize=(12, 6))

# Plot Sigmoid and its derivative
plt.subplot(1, 2, 1)
plt.plot(z, sigmoid(z), label='Sigmoid Activation', color='b')
plt.plot(z, sigmoid_grad, label="Sigmoid Derivative", color='r', linestyle='--')
plt.title('Sigmoid Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

# Plot ReLU and its derivative
plt.subplot(1, 2, 2)
plt.plot(z, relu(z), label='ReLU Activation', color='g')
plt.plot(z, relu_grad, label="ReLU Derivative", color='r', linestyle='--')
plt.title('ReLU Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

plt.tight_layout()
plt.show()
```

The above codes should help you understand the working and differences between Sigmoid and ReLU activation functions.
